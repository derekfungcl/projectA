```python
import pandas as pd
import datetime as dt
import pandas_datareader.data as web
import plotly.express as px
import plotly.graph_objects as go
start = dt.datetime(2016,1,1)
end = dt.datetime(2020,12,31)
# end = dt.datetime.now()

stocks = web.DataReader(['SPY','QQQ', 'AAPL', 'NFLX', 'GOOGL', 'MSFT'], 'yahoo', start, end)
stocks_close = pd.DataFrame(web.DataReader(['SPY','QQQ','AMZN', 'AAPL', 'NFLX', 'GOOGL', 'MSFT'], 'yahoo', start, end)['Close'])
area_chart = px.area(stocks_close.SPY, title = 'S&P PRICE (2008-2012)')

area_chart.update_xaxes(title_text = 'Date')
area_chart.update_yaxes(title_text = 'S&P', tickprefix = '$')
area_chart.update_layout(showlegend = False)

area_chart.show()
```
